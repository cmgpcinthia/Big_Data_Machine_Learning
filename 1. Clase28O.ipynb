```python
agua = {"nombre" : "brisa",
        "precio" : 100,
        "cantidad_en_stock":50,
        "importado" : True}
coke = {"nombre":"cola",
        "precio" : 300,
        "cantidad_en_stock":100,
        "importado" : False}
aguagas = {"nombre":"cristal",
        "precio" : 200,
        "cantidad_en_stock":80,
        "importado" : True}

inventario = [agua , coke , aguagas]

coke["cantidad_en_stock"] = 95

for i in inventario:
  print(i["nombre"])

valor_total = []
for i in inventario:
  totalproducto = i['precio']*i['cantidad_en_stock']
  valor_total.append(totalproducto)

print ("Valor final inventario = " , sum(valor_total))

importados = []
for i in inventario:
  if i["importado"]:
    importados.append(i["nombre"])
print ("productos importados =" , importados)
```
