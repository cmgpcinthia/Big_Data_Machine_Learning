```python
from statsmodels.tsa.seasonal import seasonal_decompose
res = seasonal_decompose(elec["OTR_PRI"])
res.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
res = seasonal_decompose(elec["INDUS_INT"])
res.plot()
```

```python
res.observed
```

```python
res.seasonal
```

```python
res.trend.plot()
```

```python
from statsmodels.tsa.seasonal import STL
res = STL(elec["INDUS_INT"], period=12).fit()
res.plot()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
hw = ExponentialSmoothing(elec["OTR_PRI"], seasonal_periods=12 , trend= "add" , seasonal= "add")
hwModel = hw.fit()
```

```python
from matplotlib import pyplot as plt
plt.plot(elec["OTR_PRI"])
plt.plot(hwModel.forecast(24))
```

##Dataset Fanta

```python
import pandas as pd
```

```python
fanta = pd.read_csv("/Ventas Mensuales Fanta 33cl.csv", sep=";" , decimal=",")
```

```python
fanta
```

```python
fanta.index = pd.to_datetime(fanta["Fecha"], format="%d/%m/%Y")
```

```python
fanta
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
res = seasonal_decompose(fanta["Unidades"], period=12)
res.plot()
```

#Herramientas de Forecasting

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
hw = ExponentialSmoothing(fanta["Unidades"], seasonal_periods=12 , trend= "add" , seasonal= "add")
hwModel = hw.fit()
```

```python
from matplotlib import pyplot as plt
plt.plot(fanta["Unidades"])
plt.plot(hwModel.forecast(12))
```

```python
hwModel.forecast(12)
```
