##Dataset Wholesale Customers

```python
customers = pd.read_csv("/content/drive/MyDrive/ESESA Lessons/Wholesale customers data.csv")
```

```python
customers
```

```python
kmeansdata = KMeans(n_clusters=4)
kmeansdata.fit(customers)
customers["clusters"] = kmeansdata.labels_
```

```python
inercias = []
for i in range(1,20):
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(customers)
  inercias.append(kmeans.inertia_)
plt.plot(inercias)
```

```python
kmeansdata = KMeans(n_clusters=4)
kmeansdata.fit(customers)
customers["clusters"] = kmeansdata.labels_
```
